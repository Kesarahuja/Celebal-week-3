### Libraries


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph


```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize=(8, 5), dpi=100)

# 2x line in light blue with triangle markers and dashed line
plt.plot(x, y, color='#ADD8E6', marker='^', linestyle='--', label='2x')

x2 = np.arange(0, 4.5, 0.5)

# X^2 lines in light pink (solid and dashed)
plt.plot(x2[:6], x2[:6]**2, color='#FFB6C1', label='X^2')
plt.plot(x2[5:], x2[5:]**2, color='#FFB6C1', linestyle='--')

plt.title('Graph', fontdict={'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0, 1, 2, 3, 4])

plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

### Bar Chart


```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
values = [1, 4, 2]

plt.figure(figsize=(5, 3), dpi=100)

bars = plt.bar(labels, values, color='lightblue')  # Set bar color to pink

patterns = ['*', '+', 'O']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)
plt.show()
```

# Dataset examples


### Line Graph - 'gas_prices.csv' 


```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in US Dollars)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, color='lightblue', marker='.', label='United States')
plt.plot(gas.Year, gas.Canada, color='yellow', marker='.', label='Canada')
plt.plot(gas.Year, gas['South Korea'], color='lightgreen', marker='.', label='South Korea')
plt.plot(gas.Year, gas.Australia, color='pink', marker='.', label='Australia')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

### Loading Data - 'fifa_data.csv'


```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(7)
```

### Histogram - 'fifa_data.csv'


```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='lightpink')

plt.xticks(bins)

plt.ylabel('No. of Players')
plt.xlabel('Levelof Skill')
plt.title('Representation of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart - 'fifa_data.csv'


```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['lightblue', '#FFC0CB']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart using ggplot - 'fifa_data.csv'


```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if isinstance(x, str) else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4, .2, 0, 0, .4)
colors = ['#FFB6C1', '#ADD8E6', '#FFDAB9', '#90EE90', '#DDA0DD']  

plt.title('Weight of Professional Soccer Players (lbs)')
plt.pie(weights, labels=labels, explode=explode, colors=colors,
        startangle=90, shadow=True, pctdistance=0.8, autopct='%.2f %%')

plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']


bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    box.set(color='lightblue', linewidth=2)

    box.set(facecolor = '#e0e0e0' )

    
plt.show()
```
